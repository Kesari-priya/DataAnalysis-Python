```python
# To Rename our columns
df = dataframe.rename(columns={'Country Name':'Country_Name',
                  'Male Height in Cm':'Male_Height_Cm',
                  'Female Height in Cm':'Female_Height_Cm',
                    'Male Height in Ft': 'Male_Height_Ft',
                       'Female Height in Ft': 'Female_Height_Ft'})
```

```python
df.columns
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
a=sns.histplot(data=df, x="Male_Height_Cm", kde=True,bins=12,color='#283747')
#b=sns.histplot(data=df, x="Male_Height_Ft")
a.set_title('Male Height (cm) Histogram',fontsize=20)
a.set_xlabel('Male Height',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
b=sns.histplot(data=df,x='Female_Height_Cm',kde=True,color='#2471A3')
b.set_title('Female Height (cm) Histogram',fontsize=20)
b.set_xlabel('Female Height',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
fig = sns.boxplot(data=df.loc[:, ["Male_Height_Ft", "Female_Height_Ft"]],orient="h",palette="Set2", linewidth=1.8)
fig.set_title("Boxplot of Male and Female Height (Ft)", fontsize=20, pad=10)
fig.set_xlabel('Height',fontsize=12)
#fig.set_ylabel="Gender", yaxis_title="Height",
plt.show()
```

```python
#plt.figure(figsize=(10,15))
sns.pairplot(df)
# to show
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["Male_Height_Cm"], label="Male Height in cm", color="#283747", kde=False)
sns.histplot(df["Female_Height_Cm"], label="Female Height in cm", kde=False)
# Add title and axes label
plt.title("Histogram of Heights of Male and Female",fontsize=20, pad=10)
plt.xlabel("Heights (cm)",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Male_Height_Ft'], label="Male Height in Ft", color="#1ABC9C", kde=False)
sns.histplot(df['Female_Height_Ft'], label="Female Height in Ft", color="#EC7063", kde=False)
# Add title and axes label
plt.title("Histogram of Heights of Male and Female",fontsize=20, pad=10)
plt.xlabel("Heights (Ft)",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,4))
Top5_male_height = df.nlargest(5, 'Male_Height_Ft')
sns.barplot(x = 'Country_Name', y = 'Male_Height_Ft', data = Top5_male_height,palette="Set1")
plt.title('Average Male Height of top 5 Countries',fontsize=20, pad=10)
plt.xticks(rotation = 25)
plt.savefig("Top 5 Average Male height.jpg")
```

```python
plt.figure(figsize=(9,4))
Top5_female_height = df.nlargest(5, 'Female_Height_Ft')
sns.barplot(x = 'Country_Name', y = 'Female_Height_Ft', data = Top5_female_height,palette="Set2")
plt.title('Average Female Height of top 5 Countries',fontsize=20, pad=10)
plt.xticks(rotation = 25)
plt.savefig("Top 5 Average Male height.jpg")
```
